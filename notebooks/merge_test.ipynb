```python
params_template = dict(
              L=9,
              T=100.0,
              dt=1.0,
              r=1,
              V="H",
              BC="0",
              E=0,
              N=3,
              totalistic=False,
              hamiltonian=False,
              trotter=True,
              symmetric=False)

Nexperiments = 2
Rs = [1,6]
ICs = ["r7", "r11"]
Qsin = np.zeros((Nexperiments, len(Rs), len(ICs)), dtype="object")
for expi in range(Nexperiments):
    for Ri, R in enumerate(Rs):
        for ICi, IC in enumerate(ICs):
            der = os.path.join(baseder, f"experiment{expi}")
            params = copy(params_template)
            params["IC"] = IC
            params["R"] = R
            Q = QCA(params, der=der)
            Q.run(tasks=["rhoj","rhojk","bisect"])
            Qsin[expi, Ri, ICi] = Q
    
```
