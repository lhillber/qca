```python
from copy import copy
import sys
sys.path.append("/../")  # look one directory up for qca modules, TODO: make a package

from qca import QCA

def reimport():
    import qca
    import networkviz
    import core
    from importlib import reload
    reload(qca)
    reload(networkviz)
    reload(core)
    from qca import QCA
reimport()
```

```python
params = dict(
              L=16,
              Lx=4,
              T=100.0,
              dt=1.0,
              r=1,
              V="H",
              BC="1-00",
              totalistic=True,
              hamiltonian=False)

evos = [{"tris": "3"}, {"blocks": "1234"}]
Rs = [2, 6]
ICs = ["f0", "f6-9"]
Es = [0.0, 0.01]
N = 100

params_list = []
for evo in evos:
    for R in Rs:
        for V in Vs:
            for IC in ICs:
                for E in Es:
                    p = copy(params)
                    if E == 0.0:
                        p["N"] = 1
                    else:
                        p["N"] = N
                    p.update(evo)
                    p["R"] = R
                    p["V"] = V
                    p["IC"] = IC
                    p["E"] = E
                    Q = QCA(p)
                    params_list.append(Q.params)
```
